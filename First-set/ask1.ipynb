```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# here we will open the image.
original_image = np.array(Image.open("trikoupi6.png"))
plt.imshow(original_image,cmap="gray")
plt.title('Original', None, 'center', None)
plt.show()
```

```python
# Show the image with a threshold value = 30
edit_image = np.copy(original_image)
threshold_value = 30

# take rows and cols of the image.
rows = edit_image.shape[0]
cols = edit_image.shape[1]

# Here i will check if is a gray photo or rgb
# GRAY has 2 values, RGB 3
if len(edit_image.shape) == 2:
    for i in range(rows):
        for j in range(cols):
            if edit_image[i][j] > threshold_value:
                edit_image[i][j] = 255
            else:
                edit_image[i][j] = 0
# End of changing the image.

plt.imshow(edit_image,cmap="gray")
plt.title("K = 30", None, 'center', None)
plt.show()
```
